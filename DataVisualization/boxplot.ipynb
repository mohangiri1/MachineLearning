# Box Plot for Visualization:

### Import necessary packages:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Data for visualization from scikit learn:

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df
```

# Use pandas to plot three different kinds of box plots:

### Single Plot:



```python
iris_df['sepal length (cm)'].plot(kind='box')
plt.show()
```

### Categorical Plot:

```python
plt.figure(figsize=(30, 10))
iris_df.boxplot(column='sepal length (cm)', by='petal width (cm)')
plt.show()
```

### Multiple Plots:

```python
plt.figure(figsize=(20, 10))
iris_df.boxplot()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
iris_df.boxplot()
plt.semilogy()
plt.show()
```

# Boxplot using seaborn:

### Single plot

```python
sns.boxplot(x = 'sepal length (cm)', data = iris_df)
plt.show()
```

```python
sns.boxplot(y = 'petal length (cm)', data = iris_df)
plt.show()
```

### Categorical Plot:

```python
sns.set(rc = {'figure.figsize': (20,10)})
sns.boxplot(y = 'sepal length (cm)', x = 'petal width (cm)', data = iris_df)
plt.show()
```

### Multiple Plots:

```python
sns.set(rc = {'figure.figsize': (20,20)})
sns.boxplot(data = iris_df)
plt.show()
```

# Boxplots Using Matplotlib:

### Single Plot:

```python
plt.figure(figsize=(10, 5))
plt.boxplot(iris_df['sepal length (cm)'])
plt.show()
```

### Categorical Plot:

```python
plt.figure(figsize=(10, 5))
plt.boxplot([iris_df['sepal length (cm)'], iris_df['petal width (cm)']], labels=['Sepal Length', 'Petal Width'])
plt.title('Box Plot of Sepal Length and Petal Width')
plt.ylabel('Value')
plt.show()
```

### Multiple Plots:

```python
plt.figure(figsize=(10, 5))
plt.boxplot([iris_df['sepal length (cm)'], iris_df['petal width (cm)'], iris_df['petal length (cm)']], labels=['Sepal Length', 'Petal Width', 'Petal Length'])
plt.title('Box Plot of Sepal Length, Petal Width, and Petal Length')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot([iris_df['sepal length (cm)'], iris_df['petal width (cm)'], iris_df['petal length (cm)']], labels=['Sepal Length', 'Petal Width', 'Petal Length'])
plt.semilogy()
plt.title('Box Plot of Sepal Length, Petal Width, and Petal Length')
plt.ylabel('Value')
plt.show()
```
