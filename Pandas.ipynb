```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,12*np.pi, 1000)
Y = np.sin(X)
plt.figure()
plt.plot(X/np.pi,Y,label = '$Sinx$')
plt.plot(X, Y-Y)
plt.legend()
plt.xlim(0,max(X)/np.pi)
```

```python
my_data = {'X':X, 'Y':Y}
df = pd.DataFrame(my_data, index = np.arange(0, len(X)))
df
```
