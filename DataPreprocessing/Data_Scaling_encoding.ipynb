# Data Scaling:

## References

*   [sklearn.preprocessing](https://scikit-learn.org/1.5/modules/preprocessing.html)

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
```

```python
df = load_iris()
df = pd.DataFrame(df.data, columns=df.feature_names)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.plot(df['sepal length (cm)'])
# plt.plot(df['sepal width (cm)'])
# plt.plot(df['petal length (cm)'])
# plt.plot(df['petal width (cm)'])
plt.title('Raw Data')
plt.show()
```

```python
df.describe()
```

```python
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(df)
scaled_df = pd.DataFrame(standard_scaled_data, columns=df.columns)
plt.plot(scaled_df['sepal length (cm)'])
# plt.plot(scaled_df['sepal width (cm)'])
# plt.plot(scaled_df['petal length (cm)'])
# plt.plot(scaled_df['petal width (cm)'])
plt.title('Standard Scaled Data')
plt.show()
scaled_df.describe()
```

```python
standard_scaler = MinMaxScaler()
standard_scaled_data = standard_scaler.fit_transform(df)
scaled_df = pd.DataFrame(standard_scaled_data, columns=df.columns)
plt.plot(scaled_df['sepal length (cm)'])
# plt.plot(scaled_df['sepal width (cm)'])
# plt.plot(scaled_df['petal length (cm)'])
# plt.plot(scaled_df['petal width (cm)'])
plt.title('MinMax Scaled Data')
plt.show()
scaled_df.describe()
```
